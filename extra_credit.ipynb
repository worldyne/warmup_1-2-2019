```python
import string

alphabet_data = [{'lower': lower, 'upper': upper, 'index':index} 
                 for lower, upper, index in zip(string.ascii_lowercase, 
                                                string.ascii_uppercase, range(1,27))]


alphabet_data
```

```python
import pandas as pd

alphabet_pd = pd.DataFrame(alphabet_data)
```
